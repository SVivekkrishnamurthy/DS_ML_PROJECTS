```python
merged_df['WMO_WIND'] =  merged_df['WMO_WIND'].fillna(merged_df['WMO_WIND'].mean())
merged_df['WMO_PRES'] =  merged_df['WMO_PRES'].fillna(merged_df['WMO_PRES'].mean())
merged_df['NEWDELHI_WIND'] = merged_df['NEWDELHI_WIND'].fillna(merged_df['NEWDELHI_WIND'].mean())
merged_df['NEWDELHI_PRES'] =  merged_df['NEWDELHI_PRES'].fillna(merged_df['NEWDELHI_PRES'].mean())
merged_df['NEWDELHI_DP'] =  merged_df['NEWDELHI_DP'].fillna(merged_df['NEWDELHI_DP'].mean())
merged_df['REUNION_WIND'] =  merged_df['REUNION_WIND'].fillna(merged_df['REUNION_WIND'].mean())
merged_df['REUNION_PRES'] =  merged_df['REUNION_PRES'].fillna(merged_df['REUNION_PRES'].mean())
merged_df['REUNION_RMW'] =  merged_df['REUNION_RMW'].fillna(merged_df['REUNION_RMW'].mean())
merged_df['LANDFALL'] =  merged_df['LANDFALL'].fillna(merged_df['LANDFALL'].mean())
```

```python
msno.bar(merged_df)
```

```python
def REU_TYPE(value):
    
    if value == '1':
        return 'D'
    elif value == '2':
        return 'DD'
    elif value == '3':
        return 'CS'
    elif value == '4':
        return 'SCS'
    elif value == '5':
        return 'EXT'
    elif value == '6':
        return 'DP'
    elif value == '7':
        return 'STC'
    elif value == '8':
        return 'OL'
    elif value == '9':
        return 'UNK'
    else: 
        return 'UNK'
    
merged_df['REUNION_TYPE'] = merged_df['REUNION_TYPE'].apply(REU_TYPE)
    
```

```python
conditions = [
    (merged_df['NEWDELHI_WIND'] < 17),
    (merged_df['NEWDELHI_WIND'] >= 17) & (merged_df['NEWDELHI_WIND'] < 28),
    (merged_df['NEWDELHI_WIND'] >= 28) & (merged_df['NEWDELHI_WIND'] < 34),
    (merged_df['NEWDELHI_WIND'] >= 34) & (merged_df['NEWDELHI_WIND'] < 48),
    (merged_df['NEWDELHI_WIND'] >= 48) & (merged_df['NEWDELHI_WIND'] < 64),
    (merged_df['NEWDELHI_WIND'] >= 64) & (merged_df['NEWDELHI_WIND'] < 120),
    (merged_df['NEWDELHI_WIND'] >= 120)
] 

GRADE = ['LPA', 'D', 'DD', 'CS', 'SVCS', 'VSCS', 'SCS']

merged_df['NEWDELHI_GRADE'] = np.select(conditions, GRADE)
```
