### Handling Categorical Variables

```python
shape=ML_Startups.shape
print("Dataset contains {} rows and {} columns".format(shape[0],shape[1]))
```

```python
ML_Startups.columns
```

```python
ML_Startups.describe()
```

```python
x=ML_Startups.iloc[:,:4]
```

```python
x
```

```python
y=ML_Startups.iloc[:,4]
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse=False)
x=ohe.fit_transform(ML_Startups[['State']])
```

```python
x
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
```

```python
col_trans=make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'),['State']),
    remainder='passthrough')
x=col_trans.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print("X_train:",x_train.shape)
print("X_test:",x_test.shape)
print("Y_train:",y_train.shape)
print("Y_test:",y_test.shape)
```

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
y_pred=linreg.predict(x_test)
y_pred
```

```python
Accuracy=r2_score(y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```

```python
plt.scatter(y_test,y_pred);
plt.xlabel('Actual');
plt.ylabel('Predicted');
```

```python
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red');
```

```python
pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
```

```python
pred_df
```
